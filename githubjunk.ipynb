```python

import pandas as pd

# Hyperlink to the CSV file
csv_link = "https://www.dropbox.com/scl/fi/yzyfwph5t1oupueluu6n0/EyeCell.csv?rlkey=dyxsq1nxhxu1tw4yncqj312wj&dl=1"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_link)

# Display the DataFrame
print(df)
# Get the data types of each column
data_types = df.dtypes

# Display the data types
print(data_types)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming df is your DataFrame with 'Year' and 'Sales' columns

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(df[['Year']], df['Sales'])

# Predict 'Sales' based on 'Year'
predictions = model.predict(df[['Year']])

# Print the coefficients
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])

# Optionally, you can print the predictions
print("Predictions:")
print(predictions)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Plot the data points
plt.scatter(df['Year'], df['Sales'], color='blue', label='Actual Sales')

# Plot the regression line
plt.plot(df['Year'], predictions, color='red', linewidth=2, label='Regression Line')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Linear Regression: Sales vs Year')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
